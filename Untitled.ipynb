```python
from scipy.sparse.linalg import expm
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = np.array([1.0,0.0],dtype=complex)
a2 = np.array([0.0,1.0],dtype=complex)
b1 = np.array([1.0,0.0],dtype=complex)
b2 = np.array([0.0,1.0],dtype=complex)

omega = a1[0]*a2[1] - a1[1]*a2[0]
z1 = 1/np.sqrt(omega)*(a1[0]+1.0j*a1[1])
z2 = 1/np.sqrt(omega)*(a2[0]+1.0j*a2[1])
```

```python
qvec = 2*np.pi*b1
gQ = np.dot(qvec,a1)*np.conj(z2)-np.dot(qvec,a2)*np.conj(z1)
gQbar = np.conj(gQ)
phi = 2*np.pi
```

```python
num_landau = 1000
toExp = np.zeros((num_landau,num_landau),dtype=complex)
```

```python
for m in range(num_landau-1):
    toExp[m][m+1] = np.sqrt(m+1)*1.0j*gQ/np.sqrt(4*np.pi)
    toExp[m+1][m] = np.sqrt(m+1)*1.0j*gQbar/np.sqrt(4*np.pi)
    
```

```python
toExp+toExp.conj().T
```

```python
zedder = expm(toExp)
```

```python
plt.imshow(np.abs(zedder))
```

```python
plt.imshow(np.abs(np.dot(zedder,zedder.conj().T)))
```
