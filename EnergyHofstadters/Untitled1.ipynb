```python
import numpy as np
import matplotlib.pyplot as plt
from math import gcd
```

```python
testMe = np.load('data_1_13.npy')
```

```python
plt.plot(testMe)
```

```python
cutoff = 150

xvals = []
totalList = []
for q in range(1,13):
    for p in range(1,q+1):
        if gcd(p, q) == 1:
            hist = np.load('rawData/data_'+str(p)+'_'+str(q)+'.npy')
            pruned_hist = [i for i in hist if abs(i) < cutoff]
            xvals = xvals + [p/q for i in range(len(pruned_hist))]
            totalList = totalList + pruned_hist
```

```python
fig = plt.figure(figsize=(8,6), dpi = 80)
plt.scatter(xvals,totalList,s=0.1,alpha=1)
plt.xlim(0,1.1)
plt.savefig('histogram_TBG_12.pdf')
```
