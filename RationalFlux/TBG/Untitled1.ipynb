```python
import numpy as np
import matplotlib.pyplot as plt
from math import gcd
```

```python
cutoff = 150

xvals = []
totalList = []
for p in range(1,4):
    for q in range(1,13):
        if gcd(p,q) == 1:
            hist = np.load('rawData/data_'+str(p)+'_'+str(q)+'.npy')
            pruned_hist = [i for i in hist if abs(i) < cutoff]
            xvals = xvals + [2*np.pi*p/q for i in range(len(pruned_hist))]
            totalList = totalList + pruned_hist
```

```python
plt.scatter(xvals,totalList,s=0.1,alpha=0.1)
plt.xlim(0,2*np.pi+1)
plt.savefig('histogram_TBG.png')
```
